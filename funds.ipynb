```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from google.colab import  drive
drive.mount('drive')
r = requests.get('http://www.gmacro.com.tw/Fund/Visitor/Fund_List.aspx?CompanyId=7d730ebb-c345-4804-816a-dd9aff22238e')
soup = BeautifulSoup(r.text,'html.parser')
# print(soup.prettify())
```

```python
davaname = soup.select("div table td a ")
dava = []
for davaname in davaname:
    davaN = davaname.text
    dava.append(davaN)
    #print(davaN)
```

```python
davavalue = soup.select("div table td span")
every = []
clearvalue = []
for everyval in davavalue:
    val = everyval.text
    every.append(val)
for num in range(0,45,3):
    clearvalue.append(every[num])
    #print(every[num])
```

```python
davaupdown = []
for num in range(1,45,3):
    davaupdown.append(every[num])
    #print(every[num])
```

```python
davaupdownpercent = []
for num in range(2,45,3):
    davaupdownpercent.append(every[num])
    #print(every[num])
```

```python
date = soup.select("div table tbody tr td")
datec = []
getdatec = []
for everyval in date:
    catch = everyval.text
    datec.append(catch)
for num in range(5,112,6):
    getdatec.append(datec[num])
#     print(datec[num])
for num in range(0,18,1):
    tex = getdatec[num]
    tex = tex.strip()
    getdatec[num] = tex
getdatec
```

```python
datec
getcoin = []
for num in range(1,105,6):
    getcoin.append(datec[num])
getcoin
for num in range(0,18,1):
    tex = getcoin[num]
    tex = tex.strip()
    getcoin[num] = tex
getcoin
```

```python
df = pd.DataFrame({'名稱':dava,'幣別':getcoin,'淨值':clearvalue,'漲跌':davaupdown,'漲跌幅':davaupdownpercent,'日期':getdatec})
df[['名稱','幣別','淨值','漲跌','漲跌幅','日期']]
df
# df.to_csv('fund.csv',index=false)
# !cp fund.csv "drive/My Drive/"
```
